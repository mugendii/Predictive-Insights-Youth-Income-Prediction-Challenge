##PRE PROCESSING

Seen from above , the columns: Matric,Degree, Diploma, Female,  and Sa citizen, have two unique vqlues 0 and 1
We change the null values to represent negative 

```python
non_nulls = train[['Matric','Degree','Diploma','Female','Sa_citizen']]
def remove_nulls(non_nulls):
  
    train[['Matric','Degree','Diploma','Female','Sa_citizen']].fillna(0, inplace=True)
    return(non_nulls)
    
```

##Next we encode the categories in province and geography to numerical labels

```python
from sklearn.preprocessing import LabelEncoder
encoded = train[['Province','Geography']]
def convert_categorical_to_numerical(train, encoded):
    # Convert categorical columns to numerical values using LabelEncoder
    encoder = LabelEncoder()
    for col in encoded:
        df[col] = encoder.fit_transform(df[col])
    return df
```

```python
convert_categorical_to_numerical(train, encoded)
```
